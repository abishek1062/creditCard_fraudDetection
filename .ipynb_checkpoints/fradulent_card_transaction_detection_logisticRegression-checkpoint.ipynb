# Severely unbalanced dataset

# Logistic Regression

```python
lR = Classifier('logisticRegression',card_data)
```

###  Method : UpSampling Minority class i.e. fraudulent transaction

```python
lR.upsampleMinorityClass()
```

#### Load trained model

```python
# lR = Classifier('loadSavedModel',card_data)
# enter ./lR_upsampled_model.sav in the textbox
```

#### Training classifier on data

```python
# Train-Test Split
lR.train_test_split()

# Training on train data 
lR.train()

# Calling method to predict labels
lR.predict_label()

# Calling method to predict probabilities
lR.predict_probability()
```

#### Checking if classifier is predicting only one label

```python
# Checking if classifier is predicting only one class
lR.check_classifier_predicts_only_one_class()
```

#### Evaluating the area under the ROC curve

```python
lR.AreaUnderCurve()
```

#### Evaluating the accuracy on the upsampled dataset

```python
lR.Accuracy()
```

#### Fetching the classification report

```python
lR.classificationReport()
```

#### Plotting the ROC curve

```python
lR.plot_ROC_curve()
```

#### Plotting the classifier vs random guessing classifier

```python
lR.plot_classifierVSrandomGuess_ROC()
```
